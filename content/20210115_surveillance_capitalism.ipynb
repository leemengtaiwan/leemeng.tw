```python
#ignore
# 實際繪圖
render_height = '750px'
network = Network(
    notebook=True, 
    height=render_height, 
    width="100%", 
    directed=True,
    heading='',
    layout=False,
)

# 定義 nodes
apps = ['谷歌' , '臉書', '推特']
nodes = {
    "使用者": {"fid": "1DcYj-O2gvG3M-uA65LmeLiB2R5s-oIi_"},
    "谷歌": {"group": "sc", "fid": "1Twnpn3tPnWQ1U1lxiQGrQEdNkuQfPnMV"},
    "臉書": {"group": "sc", "fid": "1nDZChV28WvJTGC4efiYEDbMTuK5jAKgJ"},
    "推特": {"group": "sc", "fid": "1MCYSxMfE6MWHtqY8IFnQZntux92YbBGR"},
    "政治力量": {"group": "sc", "fid": "1HKUgh9HtMdqYQmNo-xezOZbjMbqB_KwS"},
    "監控資本": {"group": "sc", "fid": "1s0ZFeDBvNOBHBZtkMX6mu3cGSk8Rt9Zw"},
    
    "監控資本主義": {"group":  "sc", "fid": "16X6Kj0-o0wWfxw1cMYv5iAVVnx617XP-"},
    "商業模式": {"group": "sc", "fid": "1tGpWGAlIYPK5Hrr3gdxFwXuZD1s-TWIe"},
    "行為預測模型": {"group": "sc", "fid": "1Xvv00YGihZmbos1S7ZxKQ9bQOlUo73aq"},
    "大他者": {"group":  "sc", "fid": "1jowkMalc5GjsU9xcWsaYDpP5FJPFdYBj"},
    "行為未來市場": {"group": "sc", "fid": "1ei0DhBFVx36MBQDgTP8hb228gE7aJNoS"},
    "行為剩餘": {"group": "sc", "fid": "159XsvGjsmHELLq_Zl7IUQTlL-oRbG1CL"},
    "全新生產手段": {"group": "sc", "fid": "1BiBTO26BoaljQBtzgWUdV25Zwc_nuZOd"},
    "監控收益": {"group": "sc", "fid": "1WgFi9hD2E2BnOYFRgzscoIg7dQNCSCss"},
    "注意力與時間": {"fid": "1J7Nrr2Z_3nkJkbHEhDotgqYq_MP1UPip"},
    
    "科技": {"group": "tech", "fid": "1Bem-_uetEmwQ3lMsQCtQgrykSYii8_7I"},
        "大數據": {"group": "tech", "fid": "1fiKgB8FjgiWbBPfgDRzlNv0RgtE8hr52"},
            "公開文本": {"group": "tech", "fid": "19M_rPCDxSOWDWv57HUBYENXgf13oYg47"},
            "影子文本": {"group": "sc", "fid": "12S1tFNxmydt43VeEImIkhI5WGICVIXE-"},
            "設計數據密集型應用": {"group": "tech", "fid": "1FgeMzjSKjTrfsig0Y0TNfHLhOdfkxB2I"},
            "物聯網": {"group": "tech", "fid": "1AWOBWlCCXl2Ifb6quVqFfLL7NcpTCbKM"},
    
    "資料科學": {"group": "tech", "fid": "1-HW1XVNKes5x0umFLkQ11lgUiRodun3G"},
        "資料分析": {"group": "tech", "fid": "1TzQCtWYeJx6e2AVQmxyZm2Qfy4BZv1wm"},
        "人工智慧": {"group": "tech", "fid": "1N_SqURgDIerhu21IisRZIjjN3kh8hiZV"},
            "機器學習": {"group": "tech", "fid": "1Ods_a_SfLaVTSmemeJyMrFl6ZjTAReP_"},
                "深度學習": {"group": "tech", "fid": "1p34c2K-Ieik9vsnPMicmIy2U0FEDBi8v"},
                    "推薦系統": {"group": "tech", "fid": "1io43NIQPwevrqEalsxfK4xLTMk7fWmFY"},
                        "點擊率預測": {"group": "tech", "fid": "1iNdWdXjYYXILY3nVR6LBLFmc9hkpoPTW"},
                    "自然語言處理" : {"group": "tech", "fid": "1H9hLcOkVu5Csuxs9sTNDl5IjBch8x9RM"},
                        "語音辨識": {"group": "tech", "fid": "1oDe8EXFcP9ClvCaVO3_zzVCe9_ngpJqP"},
                            "智慧音箱" : {"group": "tech", "fid": "1s_ZnAIIgkOkxODvVrCqGi1bT2bpIsMX-"},
                    "電腦視覺" : {"group": "tech", "fid": "1iRG_NpjA1j8T6vp4lQYD0CFbnvnvupQj"},
                        "臉部辨識" : {"group": "tech", "fid": "14Zep8RptfmK60H2a1ehEpws4GP3xMeqh"},
    
}

for node, v in nodes.items():    
    # 渲染 full gdrive url, group font color
    if 'fid' in v:
        v['shape'] = "image"
        v['image'] = gdrive_baseurl.format(fid=v['fid'])
    
    # 渲染不同 group 的 font color
    g2color = {"default": "black", "sc": "red", "tech": "blue"}
    g = v.get("group", "default")
    if 'font' not in v:
        v['font'] = {"color": g2color[g]}
    else:
        v['font']['color'] = g2color[g]
    
    # 渲染不同 group 的 node 大小
    g2value = {"default": 1, "sc": 5, "tech": 1}
    v['value'] = g2value[g]

# 微調個別 node 大小
nodes['使用者']['value'] = 3
nodes['大數據']['value'] = 3
nodes['科技']['value'] = 3
for app in apps:
    nodes[app]['value'] = 3
    
edges = [    
    Edge('監控資本主義', '大他者'),
    Edge('監控資本主義', '商業模式'),
    Edge('商業模式', '行為預測模型'),
    
    Edge("科技", "大他者"),
    Edge("科技", "全新生產手段"),
    Edge("科技", "大數據"),
    
    Edge("設計數據密集型應用", "大數據"),
    Edge("使用者", "公開文本"),
    Edge("影子文本", "行為剩餘"),
    Edge("物聯網", "大數據"),
    
    Edge("資料科學", "人工智慧"),
    Edge("資料科學", "資料分析"),
    Edge("資料分析", "行為預測模型"),
    Edge("全新生產手段", "資料科學"),
    
    Edge("人工智慧", "機器學習"),
    Edge("機器學習", "深度學習"),
    Edge("推薦系統", "點擊率預測"),
    Edge("自然語言處理", "語音辨識"),
    Edge("電腦視覺", "臉部辨識"),
    Edge("智慧音箱", "語音辨識"),
    Edge("智慧音箱", "物聯網"),
    
    Edge("行為剩餘", "全新生產手段"),
    Edge("全新生產手段", "行為預測模型"),
    Edge("點擊率預測", "行為預測模型"),
    Edge("行為預測模型", "行為未來市場"),
    Edge("行為未來市場", "監控收益"),
    Edge("使用者", "注意力與時間"),
    
    Edge("監控資本", "政治力量"),
    Edge("政治力量", "監控資本", label="合作 & 抗衡"),
]


for app in apps:
    edges.append(Edge("注意力與時間", app))
    edges.append(Edge(app, "大數據"))
    edges.append(Edge("監控收益", app))

for corpus in ['公開文本' , '影子文本']:
    edges.append(Edge('大數據', corpus))
    
for buyer in ['監控資本', '政治力量']:
    edges.append(Edge(buyer, '行為未來市場'))
    
for domain in ["推薦系統", "自然語言處理", "電腦視覺"]:
    edges.append(Edge("深度學習", domain))

for technique in ['臉部辨識', '語音辨識']:
    edges.append(Edge(technique, "影子文本"))
        
for node, v in nodes.items():
#     v['value'] = v.get('value', 1)
    network.add_node(node, **v)

for edge in edges:
    values = asdict(edge)
    render_values = dict(values)
    for k, v in values.items():
        if not v:
            render_values.pop(k)
            
    network.add_edge(**render_values)
    
# 蒐集 inbound/outbound 鄰居
neibors  = {}
for d in network.get_edges():
    src, tgt = d['from'], d['to']
    
    l = neibors.get(src, [])
    l.append(tgt)
    neibors[src] = l
    
    l2 = neibors.get(tgt, [])
    l2.append(src)
    neibors[tgt] = l2
    
# 為每個 node 加上鄰居說明
for node in network.nodes:
    neibors_text = "<br>".join([f'- {t}' for t in neibors[node['label']]])
    node['title'] = f"""與 {node["label"]} 相連實體：<br> {neibors_text}"""

network.toggle_physics(True)
# network.set_options("""""")
# network.show_buttons(filter_=['nodes', 'edges', 'physics'])
network.show(pyvis_output_file)
```
